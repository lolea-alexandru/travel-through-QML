```python
# ================= IMPORTS ================= #
import matplotlib.pyplot as plt
import numpy as np
import qutip as qt
from qutip_qip.operations import rx

# ================= CONFIG ================= #
b = qt.Bloch(figsize=[3,3])

# define starting vector
init_state = qt.basis(2, 0)
b.add_states(init_state)

qt
# get rotated
theta = np.pi / 2
R_x = rx(theta)
print(R_x)

print(R_x * init_state)

hopefully_rotated_state = R_x * init_state
b.add_states(hopefully_rotated_state)

b.render()
b.show()
```
